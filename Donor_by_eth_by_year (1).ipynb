```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/guillermo/Downloads/Donor___Donor_Ethnicity_by_Organ,_Donation_Year.csv')
df
```

```python
df.rename(columns={"Unnamed: 1": "Dates", 'Unnamed: 0': 'Organ'}, inplace = True)
df.drop(['Unnamed: 2', 'Unknown', 'American Indian/Alaska Native', 'Pacific Islander'], axis=1, inplace = True)
df = df.fillna("")
df1 = df[0:13]
df1['White'] = df1['White'].str.replace(',','').astype(int)
df1['Black'] = df1['Black'].str.replace(',','').astype(int)
df1['Hispanic'] = df1['Hispanic'].str.replace(',','').astype(int)
df1['Asian'] = df1['Asian'].str.replace(',','').astype(int)
df1['Multiracial'] = df1['Multiracial'].str.replace(',','').astype(int)
df1['All Ethnicities'] = df1['All Ethnicities'].str.replace(',', '').astype(int)
df1

```

```python
plot1 = df1.loc[2:12]
plot1.drop(['All Ethnicities', 'Organ', 'Dates'], axis = 1, inplace = True)
plot1.values
years = ['2020', '2019', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011', '2010']
```

# Plot chart of donations by ethnicity by year.

```python
fig1, ax = plt.subplots(figsize =(10,7))
eth_labels = ['White', ]
plt.plot(np.flip(years),np.flip(plot1.values), label = ['Multiracial', 'Asian', 'Hispanic', 'African American', 'White'], marker = "o")
# label the legend
plt.legend()
plt.show()
```
