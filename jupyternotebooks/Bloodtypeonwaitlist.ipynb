```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/guillermo/Downloads/Waitlist___Organ_by_ABO.csv')
df
```

```python
df.drop(['All Organs', 'Unnamed: 1', 'Heart / Lung', 'Intestine', 'Abdominal Wall', 'Head & Neck: Craniofacial', 'GU: Uterus', 'Upper Limb: Bilateral', 'Upper Limb: Unilateral', 'Face, Scalp'], axis = 1, inplace = True)
df.rename(columns={'Unnamed: 0': 'Blood Type'}, inplace = True)
df.drop([0], inplace = True)
df
```

```python
df['Kidney'] = df['Kidney'].str.replace(',','').astype(int)
df['Liver'] = df['Liver'].str.replace(',','').astype(int)
df['Heart'] = df['Heart'].str.replace(',','').astype(int)
df['Lung'] = df['Lung'].astype(int)
df['Kidney / Pancreas'] = df['Kidney / Pancreas'].astype(int)

df.columns.values
```

```python
fig1, ax = plt.subplots(figsize=(8, 8))
df.plot.bar(x = 'Blood Type', ax = ax)
```
